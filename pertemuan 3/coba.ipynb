# Aras Operasi dalam Pengolahan Citra

## 1. Aras Titik (Point Level)
Operasi di level ini bekerja pada setiap piksel secara independen dari piksel lainnya. Contoh operasi aras titik adalah **brightness adjustment** dan **thresholding**.

**Contoh:** Mengubah kecerahan gambar dengan menambahkan nilai ke setiap piksel.

---

## 2. Aras Lokal (Local Level)
Operasi di aras lokal memperhitungkan sekelompok piksel di sekitar piksel yang sedang diproses. Contoh operasi lokal adalah **filtering**, seperti blur dan edge detection.

**Contoh:** Penerapan filter Gaussian Blur untuk mengaburkan gambar.

---

## 3. Aras Global (Global Level)
Pada aras global, operasi melibatkan keseluruhan gambar. Contohnya adalah **histogram equalization** yang meratakan distribusi intensitas piksel dalam sebuah gambar.

**Contoh:** Normalisasi histogram untuk meningkatkan kontras gambar.

---

## 4. Aras Objek (Object Level)
Operasi ini melibatkan deteksi dan analisis objek di dalam gambar, seperti **object recognition** atau **segmentation**.

**Contoh:** Deteksi tepi dengan algoritma Canny atau segmentasi gambar berdasarkan thresholding.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar
image = cv2.imread('myimage.jpg', cv2.IMREAD_GRAYSCALE)

# Tambahkan kecerahan dengan menambah nilai ke setiap piksel
brightness_image = cv2.convertScaleAbs(image, alpha=1, beta=50)

# Tampilkan gambar asli dan hasil
plt.subplot(1, 2, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(1, 2, 2), plt.imshow(brightness_image, cmap='gray'), plt.title('Brightness Adjusted')
plt.show()
````

````python
# Terapkan Gaussian Blur
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

# Tampilkan gambar hasil blur
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')
plt.show()
````

````python
# Equalisasi histogram untuk meningkatkan kontras
equalized_image = cv2.equalizeHist(image)

# Tampilkan gambar setelah equalisasi
plt.imshow(equalized_image, cmap='gray')
plt.title('Histogram Equalization')
plt.show()
````

````python
# Deteksi tepi menggunakan Canny
edges = cv2.Canny(image, 100, 200)

# Tampilkan hasil deteksi tepi
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')
plt.show()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menampilkan beberapa gambar sekaligus
def display_images(images, titles, cmap='gray'):
    n = len(images)
    plt.figure(figsize=(15,5))
    for i in range(n):
        plt.subplot(1, n, i + 1)
        plt.imshow(images[i], cmap=cmap)
        plt.title(titles[i])
        plt.axis('off')
    plt.show()

# Baca gambar dalam grayscale
image = cv2.imread('myimage.jpg', cv2.IMREAD_GRAYSCALE)

# 1. Menambahkan kecerahan (brightness adjustment)
brightness_image = cv2.convertScaleAbs(image, alpha=1, beta=50)

# 2. Mengurangi kecerahan
darker_image = cv2.convertScaleAbs(image, alpha=1, beta=-50)

# 3. Meningkatkan kontras (alpha > 1)
high_contrast_image = cv2.convertScaleAbs(image, alpha=2.0, beta=0)

# 4. Mengurangi kontras (alpha < 1)
low_contrast_image = cv2.convertScaleAbs(image, alpha=0.5, beta=0)

# 5. Thresholding: Mengubah gambar menjadi biner
_, threshold_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# 6. Inversi warna (negative image)
negative_image = cv2.bitwise_not(image)

# 7. Histogram equalization untuk normalisasi kontras
equalized_image = cv2.equalizeHist(image)

# Tampilkan semua gambar hasil
images = [image, brightness_image, darker_image, high_contrast_image, low_contrast_image, threshold_image, negative_image, equalized_image]
titles = ['Original Image', 'Brightened', 'Darker', 'High Contrast', 'Low Contrast', 'Thresholded', 'Negative', 'Histogram Equalization']

# Panggil fungsi untuk menampilkan gambar
display_images(images, titles)
````




# Penjelasan Operasi Pengolahan Citra

## Fungsi `display_images`:
Fungsi ini memudahkan untuk menampilkan beberapa gambar secara berdampingan dalam satu jendela plot. Kita bisa menambahkan gambar sebanyak yang kita mau dan akan ditampilkan dalam format yang rapi.

---

## Baca Gambar:
```python
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
```
Kode ini membaca gambar dalam mode grayscale. 

---

## Brightness Adjustment:
```python
brightness_image = cv2.convertScaleAbs(image, alpha=1, beta=50)
```
Menambahkan kecerahan pada gambar dengan menambah nilai `beta` ke setiap piksel.

---

## Pengurangan Kecerahan:
```python
darker_image = cv2.convertScaleAbs(image, alpha=1, beta=-50)
```
Mengurangi kecerahan dengan nilai negatif `beta`.

---

## Penyesuaian Kontras:

- **High Contrast**:
    ```python
    high_contrast_image = cv2.convertScaleAbs(image, alpha=2.0, beta=0)
    ```
    Meningkatkan kontras dengan menggkan nilai setiap piksel (`alpha = 2`).

- **Low Contrast**:
    ```python
    low_contrast_image = cv2.convertScaleAbs(image, alpha=0.5, beta=0)
    ```
    Mengurangi kontras dengan mengalikan setiap piksel dengan faktor `alpha` yang kurang dari 1.

---

## Thresholding:
```python
_, threshold_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
```
Menerapkan thresholding biner pada gambar. Piksel di atas nilai threshold (127) diubah menjadi putih (255), dan di bawah threshold diubah menjadi hitam (0).

---

## Inversi Warna:
```python
negative_image = cv2.bitwise_not(image)
```
Melakukan inversi warna, mengubah piksel terang menjadi gelap dan sebaliknya.

---

## Histogram Equalization:
```python
equalized_image = cv2.equalizeHist(image)
```
Melakukan normalisasi kontras dengan meratakan distribusi intensitas piksel dalam gambar. Operasi ini sangat berguna untuk gambar yang kurang pencahayaan atau gambar dengan kontras yang rendah.

---

# Hasil yang Ditampilkan:
1. **Original Image**: Gambar asli dalam grayscale.

3. **Darker**: Gambar dengan kecerahan dikurangi.

5. **Low Contrast**: Gambar dengan kontras yang dikurangi.
6. **Thresholded**: Gambar yang diubah menjadi biner berdasarkan nilai threshold.
7. **Negative**: Gambar hasil inversi warna.
8. **Histogram Equalization**: Gambar dengan kontras yang dinormalisasi menggunakan histogram equalization.

````python
import cv2
import matplotlib.pyplot as plt

# Baca gambar grayscale
image = cv2.imread('myimage.jpg', cv2.IMREAD_GRAYSCALE)

# Melakukan interpolasi untuk meningkatkan resolusi gambar
hd_image = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

# Menggunakan filter kernel untuk meningkatkan ketajaman gambar (sharpening)
kernel = np.array([[0, -1, 0], 
                   [-1, 5,-1], 
                   [0, -1, 0]])
sharpened_image = cv2.filter2D(hd_image, -1, kernel)

# Menampilkan gambar sebelum dan sesudah (grayscale dan sharpened)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Grayscale')

plt.subplot(1, 2, 2)
plt.imshow(sharpened_image, cmap='gray')
plt.title('Sharpened and HD Image')

plt.show()
````

**Meningkatkan Resolusi:**
`cv2.resize()` dengan `INTER_CUBIC` untuk melakukan interpolasi bicubic, yang sering digunakan untuk memperbesar gambar dengan hasil lebih halus.
Penajaman Gambar (Sharpening):
Kernel yang digunakan (kernel) adalah filter untuk meningkatkan ketajaman gambar. Filter ini menonjolkan batasan piksel dan meningkatkan detail gambar

````python
import cv2
import matplotlib.pyplot as plt

# Baca gambar grayscale
image = cv2.imread('teshd.jpg', cv2.IMREAD_GRAYSCALE)

# Melakukan interpolasi untuk meningkatkan resolusi gambar
hd_image = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

# Menggunakan filter kernel untuk meningkatkan ketajaman gambar (sharpening)
kernel = np.array([[0, -1, 0], 
                   [-1, 5,-1], 
                   [0, -1, 0]])
sharpened_image = cv2.filter2D(hd_image, -1, kernel)

# Menampilkan gambar sebelum dan sesudah (grayscale dan sharpened)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Grayscale')

plt.subplot(1, 2, 2)
plt.imshow(sharpened_image, cmap='gray')
plt.title('Sharpened and HD Image')

plt.show()
````

````python
import cv2
import matplotlib.pyplot as plt

# Baca gambar berwarna (tidak dalam grayscale)
image = cv2.imread('teshd.jpg')

# Melakukan interpolasi untuk meningkatkan resolusi gambar
hd_image = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

# Menggunakan filter kernel untuk meningkatkan ketajaman gambar (sharpening)
kernel = np.array([[0, -1, 0], 
                   [-1, 5,-1], 
                   [0, -1, 0]])
sharpened_image = cv2.filter2D(hd_image, -1, kernel)

# Simpan gambar yang sudah ditingkatkan kualitasnya
cv2.imwrite('sharpened_hd_image.jpg', sharpened_image)

# Tampilkan gambar sebelum dan sesudah proses
plt.figure(figsize=(10, 5))

# Gambar asli
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

# Gambar yang telah diperbesar dan ditajamkan
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.title('HD and Sharpened Image')

plt.show()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar berwarna
image = cv2.imread('teshd.jpg')

# Menggunakan filter Gaussian Blur untuk menghaluskan noise
blurred_image = cv2.GaussianBlur(image, (3, 3), 0)

# Kernel untuk penajaman gambar
kernel_sharpening = np.array([[0, -1, 0], 
                              [-1, 5,-1], 
                              [0, -1, 0]])

# Aplikasikan penajaman gambar
sharpened_image = cv2.filter2D(blurred_image, -1, kernel_sharpening)

# Menggunakan CLAHE (Contrast Limited Adaptive Histogram Equalization) untuk meningkatkan kontras
lab = cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2LAB)
l, a, b = cv2.split(lab)

clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
cl = clahe.apply(l)

limg = cv2.merge((cl,a,b))
final_image = cv2.cvtColor(limg, cv2.COLOR_LAB2BGR)

# Simpan gambar yang sudah ditingkatkan kualitasnya
cv2.imwrite('enhanced_hd_image.jpg', final_image)

# Tampilkan gambar sebelum dan sesudah proses
plt.figure(figsize=(10, 5))

# Gambar asli
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

# Gambar yang telah diperbesar dan ditajamkan
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(final_image, cv2.COLOR_BGR2RGB))
plt.title('Enhanced HD Image')

plt.show()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar berwarna
image = cv2.imread('teshd.jpg')

# Terapkan Gaussian Blur untuk memisahkan detail halus
blurred_image = cv2.GaussianBlur(image, (9, 9), 10.0)

# Terapkan Unsharp Masking dengan menambahkan gambar asli dengan versi yang dikurangi blur-nya
sharpened_image = cv2.addWeighted(image, 1.5, blurred_image, -0.5, 0)

# Simpan gambar hasil yang lebih tajam
cv2.imwrite('unsharp_mask_image.jpg', sharpened_image)

# Tampilkan gambar sebelum dan sesudah proses
plt.figure(figsize=(10, 5))

# Gambar asli
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

# Gambar yang telah ditajamkan
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.title('Sharpened Image with Unsharp Mask')

plt.show()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar berwarna
image = cv2.imread('myimage.jpg')

# Perkalian citra dengan skalar (kalibrasi kecerahan)
scalar_multiply = 3
brightened_image = cv2.convertScaleAbs(image, alpha=scalar_multiply, beta=0)

# Pembagian citra dengan skalar (normalisasi kecerahan)
scalar_divide = 3
normalized_image = cv2.convertScaleAbs(image, alpha=1/scalar_divide, beta=0)

# Simpan gambar hasil
cv2.imwrite('brightened_image.jpg', brightened_image)
cv2.imwrite('normalized_image.jpg', normalized_image)

# Tampilkan gambar asli dan hasil
plt.figure(figsize=(15, 5))

# Gambar asli
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

# Gambar setelah perkalian dengan skalar
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(brightened_image, cv2.COLOR_BGR2RGB))
plt.title('Brightened Image (x3)')

# Gambar setelah pembagian dengan skalar
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(normalized_image, cv2.COLOR_BGR2RGB))
plt.title('Normalized Image (/3)')

plt.show()
````
